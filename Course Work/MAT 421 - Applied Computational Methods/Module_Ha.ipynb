# Ordinary and Partial Differential Equations

To review, let's recall that a differential equation is a relationship between some function $f(x)$ and it's derivative. When there is only a single independent variable $x$ in the function, we say that it is an ordinary differential equation. If more than one independent variable is present, such as a differential equation involving the function $g(x,y)$, then it is known as a partial differential equation.

# Euler's Method

A common way of solving ordinary differential equations is with Euler's method. Say that we have some system $F$ for which we want to find the approximate state of at point $x_{i}$. We can reference this state as an ordered pair $(x_{i}, F(x_{i})$. To perform this method in python, reference the code below:

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Define parameters
F = lambda x, f: (x**2 + 2*x + 4) # Define function
h = 0.1 # Define step length
x = np.arange(0, 1 + h, h) # Define array of x values to be tested
f_naught = 2 # Define beginning point

f = np.zeros(len(x))
f[0] = f_naught

iter_len = len(x) - 1
for i in range(0, iter_len):
      f[i + 1] = f[i] + h*F(x[i], f[i])

plt.figure(figsize = (12, 8))
plt.plot(x, f, '-o', label = 'Euler Method Solution')
plt.plot(x, (2*x + 2), 'g', label = 'Exact Solution')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.legend()
plt.show()
```
