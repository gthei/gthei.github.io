# 17.1 - Interpolation
Suppose we have a set of data points (xi, yi) where x is independent, y is dependent on x such that f(xi) = yi, and i=1,2,3,...,n. What function, then, can we create that fits** to this dataset? We know that yi is a function of xi, but how can we infer a relationship and find the function? To do this, we create a new data point, let's call it x*, and plug that value into the function so that we have f(x*).

**Note that we we say "fit", we are talking about a distinct concept from a regression. In a regression, we are attempting to find a line of best fit, while in an interpolation we are trying to find a function which directly connects the given datapoints.

# Linear Interpolation
Let's look at a basic example of interpolation involving linear functions. 

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# First we will define our set of sample data points
X = [1, 2, 3, 4] # set of sample x values
Y = [7, 8, 5, 6]  # set of sample y values

# Now we will choose a x* to interpolate
x_star = 3.5

# Next we will find a interpolation along the y-axis
f = interp1d(X, Y)
y_inter = f(x_star)

# We will first print the value to check if it is right
print(y_inter)

# Next, we will plot the interpolation
plt.style.use('seaborn-poster')
plt.figure(figsize=(10, 10))
plt.plot(X, Y, '-ob', x_star, y_inter, 'ro')
plt.title('Interpolation Points')
plt.xlabel('X-Coords')
plt.ylabel('Y-Coords')
plt.show()
```

# Cubic Spline Interpolation
Another common interpolation type is the cubic spline interpolation.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

# First let's define our datasets
X = [1, 2, 3, 4]  # set of sample x values
Y = [7, 8, 5, 6]  # set of sample y values

# As before, we will choose a x* to interpolate, in this case we choose a set of points
x_star = np.round(np.linspace(0, 10, 100), 2)

# Next, we define our cubic spline function
f = CubicSpline(X, Y)
y_inter = f(x_star)

# We then print the interpolation to check our work
print(y_inter)

# Finally we graph the points to check our work graphically
plt.style.use('seaborn-poster')
plt.figure(figsize=(10, 10))
plt.plot(x_star, y_inter, 'ro')
plt.plot(X, Y, '-ob')
plt.title('Interpolation Using Cubic Spline')
plt.xlabel('X-Coord')
plt.ylabel('Y-Coord')
plt.show()
```
