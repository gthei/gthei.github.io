# Approximate Differentiation

Finite Difference Approximation is a method of approximating derivatives of functions.

Say that we had a function $f(x)$. We would increment forward by some small change to the value of $x$, denoted by the letter $k$.

We first approximate going forward by finding the difference between $f(x)$ and $f(x+k)$. We then find the difference going backwards $f(x)$ and $f(x-k)$. We then find a line connecting $(x+k, f(x+k))$ and $(x-k, f(x-k))$.

To perform this operation in python, we will use numpy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of x and h that we will be using for f(x) and f(x±k)
h = 0.01
k = 1
x = np.arange(0, (4*np.sqrt(2)/2)*np.pi, h)

# Next we will define the function that will manipulate the given values of x
y = -np.cos(x)

# Next we will solve
f_star = np.diff(y)/h
x_star = x[:-k:]

# We will find an extact solution to compare our estimation to
f_ex = np.sin(x_star)

# Graph the solutions
plt.figure(figsize = (12,8))
plt.plot(x_star, f_star, '-o', label = 'Approximate solution')
plt.plot(x_star, f_ex, '-x', label = 'Exact solution')
plt.legend()
plt.show()
```
